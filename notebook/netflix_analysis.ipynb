# Load Dataset
In this section, the Netflix dataset is loaded into a Pandas DataFrame for analysis.

```python
# Import the Pandas library
import pandas as pd

# Read the Netflix dataset
df = pd.read_csv("../data/netflix_titles.csv")
```

Generate summary statistics for numerical and categorical columns.

# Data Cleaning
Identify and handle missing values and duplicate records.

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates()
```

# Movie vs TV Shows
This section analyzes the distribution of Movies and TV Shows available on Netflix.

```python
# Count the number of Movies and TV Shows
content_type = df["type"].value_counts()

# Display the result
content_type
```

### Analysis 
Movies represent the majority of the content available on Netflix, while TV Shows make up a smaller portion of the catalog.

# Top 10 Countries
This section identifies the countries with the highest number of Netflix titles.

```python
# Count the number of titles by country
top_countries = df["country"].value_counts().head(10)

# Display the top 10 countries
top_countries
```

### Analysis 
The results show which countries contribute the largest number of titles to the Netflix catalog.

# Top 10 Genres
This section identifies the most common genres available on Netflix.

```python
# Count the most common genres
top_genres = df["listed_in"].value_counts().head(10)

# Display the top 10 genres
top_genres
```

### Analysis

This analysis highlights the genres that appear most frequently in the Netflix library.

# Release Year Distribution
This section analyzes how Netflix content is distributed across release years.

```python
# Count titles by release year
release_years = df["release_year"].value_counts().sort_index()

# Display the result
release_years
```

### Analysis

The release year distribution helps identify trends in content production over time.

# Content Ratings

This section examines the distribution of content ratings on Netflix.

```python
# Count content ratings
ratings = df["rating"].value_counts()

# Display the result
ratings
```

### Analysis

The rating distribution provides insight into the target audience of Netflix content.

# Top Directors

This section identifies the directors with the highest number of titles on Netflix.

```python
# Count titles by director
top_directors = df["director"].value_counts().head(10)

# Display the top 10 directors
top_directors
```

### Analysis

This analysis identifies the directors who appear most frequently in the Netflix dataset.

# Movie vs TV Shows Visualization

This chart compares the number of Movies and TV Shows available on Netflix.

```python
import matplotlib.pyplot as plt

# Create a bar chart
content_type.plot(kind="bar", color=["#FF5733", "#33C1FF"])

# Add chart title
plt.title("Movies vs TV Shows on Netflix")

# Label axes
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")

# Rotate x-axis labels
plt.xticks(rotation=0)

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig("../images/movie_vs_tvshow.png", dpi=300)

# Display the chart
plt.show()



```

### Analysis

The chart clearly shows that Movies make up a larger portion of the Netflix catalog than TV Shows.

# Top 10 Countries Visualization

This chart displays the countries with the highest number of Netflix titles.

```python
# Create a horizontal bar chart
top_countries.plot(kind="barh")
plt.title("Top 10 Countries by Number of Netflix Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Country")

plt.tight_layout()

plt.savefig("../images/top_countries.png", dpi=300)

plt.show()
```

# Top 10 Genres Visualization

This chart shows the most common genres on Netflix.

```python
# Create a horizontal bar chart
top_genres.plot(kind="barh")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")

plt.tight_layout()

plt.savefig("../images/top_genres.png", dpi=300)

plt.show()
```

### Analysis

The chart shows the most common genres available on Netflix. Drama, International Movies, and Comedies are among the most represented categories, indicating that Netflix offers a wide variety of entertainment content.

# Content Ratings Visualization

This chart shows the distribution of content ratings.

```python
# Create a bar chart
ratings.plot(kind="bar")

plt.title("Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("../images/content_ratings.png", dpi=300)

plt.show()
```

### Analysis

The distribution of content ratings indicates that Netflix primarily targets mature audiences, with ratings such as TV-MA and TV-14 appearing most frequently.

# Release Year Trend

This chart illustrates the number of titles released each year.

```python
# Create a line chart

release_years.plot(kind="line")

plt.title("Number of Netflix Titles by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.tight_layout()

plt.savefig("../images/release_year_trend.png", dpi=300)

plt.show()

```

### Analysis

The number of titles generally increases over the years, reflecting Netflix's rapid expansion and investment in content production. Recent years show a significant growth in the size of the content library.
